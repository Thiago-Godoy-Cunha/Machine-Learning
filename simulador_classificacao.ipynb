```python
import numpy as np
import pandas as pd
from xgboost import XGBClassifier
```

```python
previsores = pd.read_csv("previsores.csv", header=None)
previsores = previsores.values
previsores
```

```python
alvo = pd.read_csv("alvo.csv", header=None)
alvo = alvo.values
```

```python
xg = XGBClassifier(objective="binary:logistic", n_estimators= 250, max_depth=2, learning_rate=0.05)
xg.fit(previsores, alvo)
paciente1 = np.array([40, 0, 1, 140, 289, 0, 0, 172, 0, 0, 0]).reshape(1, -1)
paciente2 = np.array([48, 1, 3, 138, 214, 0, 0, 108, 1, 1.5, 1]).reshape(1, -1)
resultado = xg.predict(paciente2)
resultado
```
