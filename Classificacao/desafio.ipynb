```python
import pandas as pd
import numpy as np  
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split, KFold, cross_validate
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("data_cancer2.csv")
df = df.iloc[:, 0:32]
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
X.replace({col: 0 for col in X.columns}, np.nan, inplace=True)
X.fillna({col: X[col].mean() for col in X.columns}, inplace=True)
X = MinMaxScaler().fit_transform(X)
y = LabelEncoder().fit_transform(y)
```

```python
kfold = KFold(n_splits=20, shuffle=True, random_state=0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
xgb = XGBClassifier(max_depth=2, learning_rate=0.05, n_estimators=250, random_state=0)
xgb.fit(X_train, y_train)
previsoes_xgb = xgb.predict(X_test)
modelo = GaussianNB()
resultado = cross_validate(xgb, X_train, y_train, cv = kfold, return_train_score=True)

print("Acurácia de Treino (Média): ", resultado['train_score'].mean())
print("Acurácia de Validação cruzada (Média): ", resultado['test_score'].mean())

print("Acurácia com dados de teste: %.2f%%" % (accuracy_score(y_test, previsoes_xgb) * 100))

```
